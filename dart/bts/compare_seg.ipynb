```python
import nibabel as nb
import numpy as np
import matplotlib.pyplot as plt
```

```python
brats = nb.load('/Users/sravan953/Documents/CU/Projects/imr-framework/DART/Data/MICCAI_BraTS_2018_Data_Training/Nifti/HGG/Brats18_2013_2_1/Brats18_2013_2_1_t1.nii.gz')
brats = brats.get_fdata()
print(f'BraTS: {brats.shape}')

brats_registered = nb.load('/Users/sravan953/Documents/CU/Projects/imr-framework/DART/Data/brats_registered.nii')
brats_registered = brats_registered.get_fdata()
print(f'BraTS: {brats_registered.shape}')

atlas = np.load('/Users/sravan953/Documents/CU/Projects/imr-framework/DART/Data/atlas_resampled_oriented.npy')
# atlas = atlas.get_fdata()
print(f'Atlas: {atlas.shape}')

atlas_seg = np.load('/Users/sravan953/Documents/CU/Projects/imr-framework/DART/Data/atlas_seg_resampled_oriented.npy')
# atlas_seg = atlas_seg.get_fdata()
print(f'Atlas segmented: {atlas_seg.shape}')
```

```python
slice = 79

def plot(slice):
    slice = np.rot90(slice, k=1)
    slice = np.fliplr(slice)
    plt.imshow(slice)

plt.figure()
plot(brats[:, slice, :])
plt.title('BraTS')

plt.figure()
plot(brats_registered[:, slice, :])
plt.title('BraTS registered to atlas')

plt.figure()
plot(atlas[:, slice, :])
plt.title('Atlas')

plt.figure()
plot(atlas_seg[:, slice, :])
plt.title('Atlas segmented')
```
